```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Function to display frame inline in Jupyter notebook
def display_frame(frame):
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('on')
    plt.show()

def select_roi(frame):
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.title("Click 2 points to form a rectangle (top-left and bottom-right)")
    
    # Use ginput to get 2 points from the user
    points = plt.ginput(2)
    plt.close()
    
    # Extract the coordinates
    x1, y1 = int(points[0][0]), int(points[0][1])
    x2, y2 = int(points[1][0]), int(points[1][1])
    
    # Make sure x1, y1 is the top-left and x2, y2 is the bottom-right
    roi = (min(x1, x2), min(y1, y2), abs(x2 - x1), abs(y2 - y1))
    return roi
```

```python
# Load video
cap = cv2.VideoCapture('Robots.mp4')

# Read first frame
ret, frame = cap.read()
gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Initialize SIFT detector
sift = cv2.SIFT_create()

display_frame(frame)
```
